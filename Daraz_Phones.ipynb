For 1 page

```python
data=[]
all_phones= data_json['mods']['listItems']
for item,phone in enumerate(all_phones):
    
    try:
        originalPrice = phone['originalPrice']
    except:
        originalPrice = phone['price']
    try:
        discount = phone['discount']
    except:
        discount = 'Not available'
    try:
        ratingScore = phone['ratingScore']
    except:
        ratingScore = 'Not available'
        
    try:
        review = phone['review']
    except:
        review = 'Not available'
    
    
    
    
    d={
        
        'name' : phone['name'],
        'originalPrice' : originalPrice,
        'discount' : discount,
        'ratingScore' : ratingScore,
        'review' : review,
        'brandName' : phone['brandName']
    
        
    }
    
    
    data.append(d)
    
```

```python
df=pd.DataFrame(data)
df
```

Now for all the pages

```python
data=[]
for page in range(1,12):
    url= f'https://www.daraz.com.bd/smartphones/?ajax=true&page={page}&spm=a2a0e.home.cate_6.1.735212f79063zo&style=wf'
    response = requests.get(url)
    data_json = response.json()
    print(f'{page} no page is done')
    all_phones= data_json['mods']['listItems']
    for item,phone in enumerate(all_phones):

        try:
            originalPrice = phone['originalPrice']
        except:
            originalPrice = phone['price']
        try:
            discount = phone['discount']
        except:
            discount = 'Not available'
        try:
            ratingScore = phone['ratingScore']
        except:
            ratingScore = 'Not available'

        try:
            review = phone['review']
        except:
            review = 'Not available'




        d={

            'name' : phone['name'],
            'originalPrice' : originalPrice,
            'discount' : discount,
            'ratingScore' : ratingScore,
            'review' : review,
            'brandName' : phone['brandName']


        }


        data.append(d)
    
    
    
    
```

```python
len(data)
```

```python
df=pd.DataFrame(data)
df
```

```python
df.to_csv('Daraz_phones.csv',index=False)
```
